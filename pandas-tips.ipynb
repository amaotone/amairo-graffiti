```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import display
```

```python
df = pd.read_csv('input/train.csv', index_col=0)
```

```python
print(df.shape)
display(df.head())
```

```python
new = df.query("Pclass==1 and Age >= 30")
print(new.shape)
display(new.head())
```

```python
new = df.filter(regex='^P')
print(new.shape)
display(new.head())
```

```python
new = df.select_dtypes('object')
print(new.shape)
display(new.head())
```

```python
new = df.select_dtypes(exclude=np.float)
print(new.shape)
display(new.head())
```

```python
display(df.groupby('Pclass')['Survived'].value_counts().unstack())
```

```python
display(df.pivot_table(index='Sex', columns='Pclass', values='Survived', aggfunc=np.mean))
```
